## Match Results

```python
# importing necessary libraries

import pandas as pd
import json
```

```python
pd.set_option('display.max_rows',1000)
pd.set_option('display.max_columns',20)
```

```python
with open('./t20_json_files/t20_wc_match_results.json') as f:
    data = json.load(f)
data[0]['matchSummary']
```

```python
df_match_results = pd.DataFrame(data[0]['matchSummary'])
df_match_results
```

```python
df_match_results.shape
```

```python
df_match_results.rename({'scorecard':'match_id'},axis = 1,inplace = True)
```

```python
df_match_results.head()
```

```python
match_id_dict = {}

for index,row in df_match_results.iterrows():
    key1 = row.team1 + ' Vs ' + row.team2
    key2 = row.team2 + ' Vs ' + row.team1
    match_id_dict[key1] = row.match_id
    match_id_dict[key2] = row.match_id
match_id_dict
```

```python
df_match_results.to_csv('t20_csv_files/match_results.csv',index = False)
```

## Batting Summary

```python
with open('./t20_json_files/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['battingSummary'])

all_records
```

```python
df_batting_summary = pd.DataFrame(all_records)
df_batting_summary.head(10)
```

```python
df_batting_summary['out/not_out'] = df_batting_summary.dismissal.apply(lambda x: 'out' if len(x)>0 else 'not_out')
df_batting_summary
```

```python
df_batting_summary.drop(columns = 'dismissal', inplace = True)
```

```python
df_batting_summary.batsmanName.replace('â€','',regex=True, inplace=True)
```

```python
df_batting_summary['batsmanName'] = df_batting_summary.batsmanName.str.strip()
df_batting_summary
```

```python
df_batting_summary['match_id'] = df_batting_summary.match.map(match_id_dict)
df_batting_summary
```

```python
df_batting_summary.to_csv('t20_csv_files/batting_summary.csv',index=False)
```
